```python
# Script to convert Amsterdam dialect excel file to RDF 
# marieke.van.erp@dh.huc.knaw.nl
# date: 4 February 2019

# To do: figure out how to read in filemaker files 

import pandas as pd
import numpy as np

data.empty

data = pd.read_excel('Amsterdamsedata21-2-2019_copy.xls')#.iloc[:20]

# Turn empty values into empty strings
data = data.replace(np.nan, '', regex=True)
data = data.replace(" $", "", regex=True)

# Add identifiers for each term 
data = data.reset_index()
data['New_ID'] = data.index + 1
```

```python
# First clean up cells and split where necessary 

def pir(df, c):
    colc = df[c].str.split(',|;|\/')
    clst = colc.values.tolist()
    lens = [len(l) for l in clst]
    cdf = pd.DataFrame({c: np.concatenate(clst)}, df.index.repeat(lens))
    return df.drop(c, 1).join(cdf).reset_index(drop=True)

```

```python
# Split 'jaartal' and convert to owltime value 
data = pir(data, 'jaartal')
```

```python
# This is where the conversion to RDF starts 

from rdflib import URIRef, BNode, Literal, Namespace, Graph, RDF, RDFS, XSD, util
import urllib.parse

# Initialise graph 
g = Graph()

# Define relevant namespaces 
atm = Namespace("http://amsterdamtimemachine.nl/lexdata/")
g.bind('atm', atm)

time = Namespace("http://www.w3.org/2006/time#")
g.bind('time', time)

lemon = Namespace("http://lemon-model.net/lemon#")
g.bind('lemon', lemon)

lexinfo = Namespace("http://www.lexinfo.net/ontology/2.0/lexinfo#")
g.bind('lexinfo', lexinfo)

prov = Namespace("http://www.w3.org/ns/prov#")
g.bind('prov', prov)

dc = Namespace("http://purl.org/dc/elements/1.1/")
g.bind('dc', dc)

#penn = Namespace("http://purl.org/olia/penn.owl#")
#olia = Namespace("http://purl.org/olia/olia.owl#")
#xsd = Namespace("http://www.w3.org/2001/XMLSchema#")
#skos = Namespace("http://www.w3.org/2004/02/skos#")
#dct = Namespace("http://purl.org/dc/terms/")
#dbo = Namespace("http://dbpedia.org/ontology/")
#adamlink = Namespace("http://adamlink.nl/geo")

# Create ontolex Lexical Entries via identifiers from 'New_ID'
import urllib.parse


def add_atm_ns_int(cell):
    newcell = atm + str(cell)
    return newcell 

def add_atm_ns_str(cell):
    newcell = ''
    if len(cell) > 1:
        newcell = atm + cell
    return newcell

atmlexicon = URIRef(atm + "ATMLexicon")
lemonlexicon = URIRef(lemon + "Lexicon")
lemonlanguage = URIRef(lemon + "language")
g.add(( atmlexicon, RDF.type, lemonlexicon ))
g.add(( atmlexicon, lemonlanguage, Literal("nl")))

for index, row in data.iterrows():
    # Generate lexical entries from column "New_ID"
    item = add_atm_ns_int(row["New_ID"])
    lexicalEntry = URIRef(item)
    g.add((atmlexicon, URIRef(lemon + "entry"), lexicalentry))
    g.add((lexicalEntry, RDF.type, URIRef(lemon + "LexicalEntry")))
    # Add lemon:Form from column "Amsterdams"
    amsterdams = Literal(row["Amsterdams"])
    lexicalForm = BNode()
    g.add((lexicalEntry, URIRef(lemon + "otherForm"), lexicalForm ))
    g.add((lexicalForm, URIRef(lemon + "writtenRep"), amsterdams ))
    # Add lemon:sense 
    lexicalSense = BNode()
    g.add((lexicalEntry, URIRef(lemon + "sense"), lexicalSense))
    # Add meaning 
    # To do: where does the meaning come from? 
    meaning = Literal(row["betekenisnieuw"])
    g.add( (lexicalSense, URIRef(lemon + "reference"), meaning ) )
    # Add 'Dutch translation' from column "Nederlands"
    ###############################
    ###############################
    ### Modelled as canonicalForm instead of translation 
    #dutchEntry = BNode()
    #lexicalForm2 = BNode()
    #lexicalSense2 = BNode()
    #g.add((lexicalSense, URIRef(lemon + "translationOf"), lexicalSense2))
    #g.add((dutchEntry, RDF.type,  URIRef(lemon + "LexicalEntry")))
    #g.add((dutchEntry, URIRef(lemon + "sense"), lexicalSense2))
    #g.add((dutchEntry, URIRef(lemon + "form"), lexicalForm2))
    #g.add((lexicalForm2, URIRef(lemon + "writtenRep"), Literal(row["Nederlands"])))
    ###############################
    ###############################
    lexicalForm2 = BNode()
    g.add((lexicalEntry, URIRef(lemon + "canonicalform"), lexicalForm2))
    g.add((lexicalForm2, URIRef(lemon + "writtenRep"), Literal(row["Nederlands"])))
    # Add etymology from column "herkomst"
    # Note: check if this can be a literal or should be modelled differently 
    if pd.isnull(row["herkomst"]) == False:
        g.add((lexicalSense, URIRef(lexinfo + "etymology"), Literal(row["herkomst"])))
    # Add lemon:context 
    lexicalContext = BNode()
    g.add((lexicalSense, URIRef(lemon + "context"), lexicalContext))
    # Add geographical location (neighbourhood)
    if pd.isnull(row["dialectRegion"]) == False:  
        location = URIRef(row["dialectRegion"])
        g.add((lexicalContext, URIRef(dc + "coverage"), location))
        g.add( (location, RDF.type, URIRef(dc + "Location")) )
    # Source
    # Create a source instance via prov-o
    source = BNode()
    g.add( (source, RDF.type, URIRef(prov + "Entity") ))
    g.add((lexicalContext, URIRef(prov + "wasDerivedFrom"), source))
    g.add( ( source, URIRef(dc + "title"), Literal(row["titel"]) ) )
    g.add( ( source, URIRef(dc + "creator"), Literal(row["auteur"]) ) )
    g.add( ( source, URIRef(dc + "publisher"), Literal(row["uitgeverij"]) ) )
    # Create a year instance
    year = Literal(row["jaartal"], datatype=XSD.integer)
    date = BNode()
    g.add(( date, RDF.type, URIRef(time + "DateTimeDescription") ))
    g.add( (date, URIRef(time + "year"), year) )
    # Connect year to source 
    g.add( (source, URIRef(dc + "date"), date) )
    # To do: get author names and publication title from 'bibliografische gegevens' field
    # Add informant 
    if pd.isnull(row['informant of bron']) == False:
        informant = URIRef(atm + urllib.parse.quote(row['informant of bron']))
        informantName = Literal(row['informant of bron'])
        g.add( (informant, RDFS.label, informantName) )
        g.add( (informant, RDF.type, URIRef(prov + "Entity")) )
        g.add( (lexicalForm, URIRef(prov + "wasAttributedTo"), informant ) )
    # Add sociolect
    if pd.isnull(row['sociolect']) == False:
        sociolect = Literal(row['sociolect'])
        g.add( (lexicalContext, URIRef(lemon + "usage"), sociolect) )

bl = g.serialize(format='pretty-xml')

#print(bl.decode())
```
